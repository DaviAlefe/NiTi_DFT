```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```
# ordem dos parametros: [A,B,C,cosAB]
param_opt = [2.88, 4.6, 4.1, np.cos(np.deg2rad(96))]
dparam = [0.005, 0.05, 0.05, 0.01]
```

```
# parametros deslocados

A_d = [param_opt[0]+dparam[0],param_opt[0]-dparam[0]]
B_d = [param_opt[1]+dparam[1],param_opt[1]-dparam[1]]
C_d = [param_opt[2]+dparam[2],param_opt[2]-dparam[2]]
cosAB_d = [param_opt[3]+dparam[3],param_opt[3]-dparam[3]]
```

```
param_init = []
for iA in range(2):
  for iB in range(2):
    for iC in range(2):
      for icos in range(2):
        param_init.append([A_d[iA], B_d[iB], C_d[iC], cosAB_d[icos]])
param_init
```

```
import numpy as np

np.rad2deg(np.arccos(-0.11452846326765355))
```

```
#loop para calcular as energias dos parametros iniciais
import quantum_macchiato as qm

E = []
data = []
for i in range(len(param_init)+1):
  #altera os parametros no arquivo de entrada
  qm.chg_param('niti.monoclinic.scf.paw.in','A',param_init[i][0])
  qm.chg_param('niti.monoclinic.scf.paw.in','B',param_init[i][1])
  qm.chg_param('niti.monoclinic.scf.paw.in','C',param_init[i][2])
  qm.chg_param('niti.monoclinic.scf.paw.in','cosAB',param_init[i][3])
  #roda o arquivo com os parametros alterados
  qm.run('niti.monoclinic.scf.paw.in',nproc=32)
  #coleta a energia do arquivo de saída
  E.append(qm.collec_totE('niti.monoclinic.scf.paw.out'))
  #cria uma linha com os valores envolvidos e coloca essa linha em uma lista
  line = [param_init[i][0], param_init[i][1], param_init[i][2], param_init[i][3],E[i]]
  data.append(line)
  #escreve os resultados em um csv:
  from csv import writer
  with open('niti.monoclinic.minim.csv','w') as output:
    w = writer(output)
    w.writerow(['A [angstrom]','B [angstrom]','C [angstrom]', 'cosAB', 'Total Energy [eV]']) #escreve o header para o arquivo de saída
    for line in data:
      w.writerow(line)
```
